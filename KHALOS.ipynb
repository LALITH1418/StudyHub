```python
data=data.dropna(axis=0)
data
```

```python
data[' DATE1']=data[' DATE1'].astype('datetime64')
```

```python
data
```

```python
data['%change']=((data[' CLOSE_PRICE']-data[' PREV_CLOSE'])/data[' PREV_CLOSE'])*100
```

```python
data
```
